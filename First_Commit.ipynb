```python
class Patient : 
    def __init__(self, name, mrn, age, hemoglobin, wbc, medications : list):
        self.name = name
        self.mrn = mrn
        self.age = age
        self.hemoglobin = hemoglobin
        self.wbc = wbc
        self.medications = medications
    def is_anemic(self):
        if self.hemoglobin < 10:
            return True
        else:
            return False
    def has_abnormal_wbc(self):
        if self.wbc < 4.0 or self.wbc > 11.0:
            return True
        else:
            return False
    def get_alerts(self):
        alerts = []
        if self.is_anemic():
            alerts.append("Patient is anemic")
        if self.has_abnormal_wbc():
            alerts.append("Patient has abnormal WBC count")
        return alerts
    def summary(self):
        return f"Patient {self.name} (MRN: {self.mrn}) is {self.age} years old with hemoglobin {self.hemoglobin} g/dL and WBC {self.wbc} x10^9/L. Medications: {', '.join(self.medications)}"
```

```python
patient_1 = Patient("Sara Al-Ahmad",  "MRN-001", 34,  11.8, 4.5,  ["Methotrexate", "Folic Acid"])
patient_2 = Patient("Omar Khalil",    "MRN-002", 52, 13.9, 14.2, ["Rituximab", "Ondansetron"])
patient_3 = Patient("Lina Haddad",    "MRN-003", 28,  6.3,  2.1, ["Doxorubicin", "Cyclophosphamide", "Dexamethasone"])
patient_4 = Patient("Yousef Nasser",  "MRN-004", 61, 15.1,  7.8, ["Ibuprofen"])
patient_5 = Patient("Rania Barakat",  "MRN-005", 45,  9.7, 32.4, ["Vincristine", "Prednisone"])
```

```python
for p in [patient_1, patient_2, patient_3, patient_4, patient_5]:
    print(p.summary())
    print("Alerts:", p.get_alerts())
    print()
```
